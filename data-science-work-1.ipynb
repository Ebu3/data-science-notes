## 2.3 Nested Function

```python
def square():
    """ return square of vaule """
    def add():
        """ add two local variable """
        x = 2 
        y = 3 
        z = x + y
        return z
    return add()**2
print(square())
```

## 2.4 Default and Flexible Arguments

```python
# default arguments
def f(a,b=1,c=2):
    y = a + b + c
    return y
print(f(5))
print(f(5,4,3))
```

```python
# flexible arguments *args

def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4,5)
print("")
# flexible arguments **kwargs that is a dictionary argument

def f(**kwargs):
    """ print key and value of dictionary """
    for key, value in kwargs.items():
        print(key," ",value)
f(country = "spain",capital="madrid",population=12341)
```

## 2.5 Lambda Function

```python
# user defined function (long way)
def square(x):
    return x**2
print(square(5))
print("")
# lambda function (short way)
square = lambda x : x ** 2
print(square(4))
print("")
tot = lambda x,y,z : x+y+z
print(tot(1,3,4))
```

## Resampling pandas time series
+ Resampling : statistical method over different time intervals
 + Needs string to specify frequency like "M" = month or "A" = year
+ Downsampling : reduce date time rows to slower frequency like from daily to hourly
+ Upsampling : increase date time rows to faster frequency like from daily to hourly
+ Interpolate : Interpolate values according to different methods like "linear" "time" or "index"

```python
data2.resample("A").mean()
```

```python
data2.resample("M").mean()
```

```python
data2.resample("M").mean().interpolate("linear")
```
